## Q(Lag(W,Y)) vs Lag(W, Q(Y))

```python
import mapclassify.api as mc
np.random.seed(12089)
y = np.random.normal(size=(25,1))
y.shape
```

```python
qy = mc.Quantiles(y, k=5)
qy
```

```python
qy = qy.yb
```

```python
qy
```

```python
wqy = lps.lag_categorical(w, qy)
```

```python
wqy
```

```python
w.neighbors[24]
```

```python
qy[w.neighbors[24]]
```

```python

qy[24]
```

```python
w = lps.lat2W(5,5)
np.random.seed(12345)
for i in range(10):
    y = np.random.normal(size=(25,1))
    yq = mc.Quantiles(y, k=4).yb
    wyq = lps.lag_categorical(w, yq, ties='random')
    for j in range(25):
        if wyq[j] not in yq[w.neighbors[j]] and wyq[j] != yq[j]:
            print(i,j)
            print(wyq[j],yq[w.neighbors[j]], yq[j])
            print("\n")
```

```python
wqy = np.array([lps.lag_categorical(w, mc.Quantiles(yi).yb) for yi in y.T]).T
```
